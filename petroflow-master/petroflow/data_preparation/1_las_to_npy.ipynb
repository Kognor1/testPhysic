# Конвертация каротажей из формата .las в .npy

```python
import os
import glob
import json

import numpy as np
import pandas as pd
import lasio
```

Определение пути к директории с исходными файлами в формате `.las`:

```python
WELL_LAS_PATH = "path/to/las"
```

Определение пути к директории с преобразованными файлами в формате `.npy`:

```python
WELL_NPY_PATH = "path/to/npy"

if not os.path.exists(WELL_NPY_PATH):
    os.makedirs(WELL_NPY_PATH, exist_ok=True)
```

Определение пути к словарю мнемоник и его загрузка:

```python
MNEMONICS_DICT_PATH = "./mnemonics.json"

with open(MNEMONICS_DICT_PATH) as file:
    mnemonics_transform_dict = json.load(file)
```

Определение пути к файлу с результатами интерпретации:

```python
INTERPRETATION_PATH = "path/to/interpretation"
```

Всего в файле с интерпретацией содержится информация о 856 уникальных скважинах:

```python
interpretation_df = pd.read_csv(INTERPRETATION_PATH, delimiter=";", encoding="cp1251")
unique_wells = interpretation_df["Номер скважины"].unique()
len(unique_wells)
```

`.las`-файлы доступны для 417 скважин:

```python
well_names = glob.glob(os.path.join(WELL_LAS_PATH, "*.las"))
well_names = [os.path.splitext(os.path.basename(name))[0] for name in well_names]
len(well_names)
```

Из них проинтерпретированными являются 411:

```python
well_names = np.intersect1d(unique_wells, well_names)
len(well_names)
```

Конвертация файлов:

```python
def read_las(path, mnemonics_transform_dict=None):
    las = lasio.read(path)
    if mnemonics_transform_dict is not None:
        for curve in las.curves:
            mnemonic = curve.original_mnemonic
            curve.mnemonic = mnemonics_transform_dict.get(mnemonic, mnemonic)
    return las
```

```python
%%time

for well_name in well_names:
    log_path = os.path.join(WELL_LAS_PATH, well_name + ".las")
    las = read_las(log_path, mnemonics_transform_dict)

    data = las.data.T
    data_path = os.path.join(WELL_NPY_PATH, well_name + "_data")
    np.save(data_path, data)

    meta = np.array(las.keys())
    meta_path = os.path.join(WELL_NPY_PATH, well_name + "_meta")
    np.save(meta_path, meta)
```
